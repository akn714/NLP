```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
paragraph = open('dune_Messiah.txt', 'r', encoding='utf-8').read()
sentences = nltk.sent_tokenize(paragraph)
```

```python
sentences
```

```python
import re

corpus = []
for document in sentences:
    document = re.sub('[^a-zA-Z]', ' ', document).lower()
    document = [word for word in document.split() if word not in stopwords.words('english')]
    corpus.append(' '.join(document))
```

```python
corpus
```

```python
# BAG OF WORDS

from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(binary=True, ngram_range=(3,100))
X = cv.fit_transform(corpus).toarray()

X
```

```python
with open('bow_dune_messiah.txt', 'w', encoding='utf-8') as f:
    for row in X:
        f.write(' '.join(map(str, row)) + '\n')
```

```python
X.shape
```

```python
# TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(ngram_range=(3,100), max_features=5000)

X_tfidf = tfidf.fit_transform(corpus).toarray()
```

```python
X_tfidf
```

```python
with open('tfidf_dune_messiah.txt', 'w', encoding='utf-8') as f:
    for row in X_tfidf:
        f.write(' '.join(map(str, row)) + '\n')
```
